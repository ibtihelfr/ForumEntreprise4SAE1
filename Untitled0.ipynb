```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Charger le dataset
df = pd.read_csv("/content/data-Tcspcng0FWQUkCPcTwBwo.csv", encoding='latin1')

# Supprimer les colonnes inutiles
columns_to_drop = ["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"]


# Renommer les colonnes
new_column_names = {"v1":"Category","v2":"Message"}
df.rename(columns=new_column_names, inplace=True)

# Convertir les catégories en valeurs numériques
df['Category'] = df['Category'].map({'ham': 0, 'spam': 1})

# Diviser les données en ensembles d'entraînement et de test
X = df["Message"]
Y = df["Category"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Extraction de fonctionnalités TF-IDF
vectorizer = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = vectorizer.fit_transform(X_train)
X_test_features = vectorizer.transform(X_test)

# Modèle de régression logistique
logistic_model = LogisticRegression()
logistic_model.fit(X_train_features, Y_train)
predict_train_logistic = logistic_model.predict(X_train_features)
predict_test_logistic = logistic_model.predict(X_test_features)

# Accuracy de la régression logistique
accuracy_train_logistic = accuracy_score(Y_train, predict_train_logistic)
accuracy_test_logistic = accuracy_score(Y_test, predict_test_logistic)

print("Accuracy on training data (Logistic Regression): ", accuracy_train_logistic)
print("Accuracy on test data (Logistic Regression): ", accuracy_test_logistic)

# Modèle SVM
svm_model = SVC(kernel='linear')
svm_model.fit(X_train_features, Y_train)
predict_train_svm = svm_model.predict(X_train_features)
predict_test_svm = svm_model.predict(X_test_features)

# Accuracy de SVM
accuracy_train_svm = accuracy_score(Y_train, predict_train_svm)
accuracy_test_svm = accuracy_score(Y_test, predict_test_svm)

print("Accuracy on training data (SVM): ", accuracy_train_svm)
print("Accuracy on test data (SVM): ", accuracy_test_svm)
# Matrice de confusion pour la régression logistique
conf_matrix_logistic = confusion_matrix(Y_test, predict_test_logistic)
plt.figure(figsize=(6,4))
sns.heatmap(conf_matrix_logistic, annot=True, fmt="d", cmap="Oranges", cbar=False)
plt.xlabel("Predicted value")
plt.ylabel("Actual value")
plt.title("Confusion Matrix (Logistic Regression)")
plt.show()

# Matrice de confusion pour SVM
conf_matrix_svm = confusion_matrix(Y_test, predict_test_svm)
plt.figure(figsize=(6,4))
sns.heatmap(conf_matrix_svm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel("Predicted value")
plt.ylabel("Actual value")
plt.title("Confusion Matrix (SVM)")
plt.show()

# Visualisation des mots les plus courants dans les e-mails spam
stop_words = set(stopwords.words('english'))
spam_words = " ".join(df[df['Category'] == 0]['Message']).split()
spam_word_freq = Counter([word.lower() for word in spam_words if word.lower() not in stop_words and word.isalpha()])

plt.figure(figsize=(10, 6))
plt.bar(*zip(*spam_word_freq.most_common(7)), color='y')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 7 Most Common Words in Spam Emails')
plt.xticks(rotation=45)
plt.show()

```
